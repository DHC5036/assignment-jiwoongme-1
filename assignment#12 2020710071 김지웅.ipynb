### What kind of application can be possible using GAN in medical domain?

의료 이미지 영역에서 GAN이 사용될 수 있을 것 같다. 이미 딥러닝이 의료 이미지 분야에서 많은 두각을 나타내고 있다.
독일 암센터 연구팀은 MRI에서 진행성 전립선 암에 대해 전문가의 병변 표식과 GAN이 생성한 병변 표식을 구분하는 모델을 만들고 그 결과를 다시 분할 모델에 학습시켜 모델의 병변 표식 성능을 기존 방식보다 유의하게 개선시킨 연구가 있다.
또 다른 연구에서는 GAN을 통해 합성 망막 이미지를 생성하는 방법을 개발하였는데 이를 통해 실제 존재하지 않는 고품질의 망막 이미지를 생성함으로써 AI 모델을 훈련하는데 사용할 수 있었다고 한다.
GAN을 의료 분야, 특히 이미지 관련 분야에서 다양하게 합성, 분할 등에 쓸 수 있을거같다.
