### 1. What happens if you implement only parts of a GRU, e.g., with only a reset gate or only an update gate?

reset gate나 update gate 중 하나만 구현이 되어있다면 현시점의 정보 후보군을 계산하는 Candidate 단계에서는 reset gate의 결과를 곱하여야하고, 은닉층 계산하는 단계에서도 update gate 결과와 candidate 결과를 결합해야하는데, 제대로 계산이 되지않아 학습이 제대로 되지 않을 것이다.
