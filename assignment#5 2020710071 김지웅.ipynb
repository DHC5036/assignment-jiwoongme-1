### 1. Assume that you want to balance a (real) ball on a (real) saddle

#### Why is this hard? 

특정 부분에서만 순간기울기가 0이기 때문에 평행을 맞추는 것이 쉽지 않다

#### Can you exploit this effect also for optimization algorithms? 

최적화 알고리즘에서 loss function의 최소값을 찾아야하는데, 현실에서의 loss fouction은 2차함수처럼  깔끔하지 않는 경우가 많고, saddle point주변이 평평하여 더디게 진행될 수도 있다. local minium이나 saddle point에 안주하지 않도록 해야한다.

### 2. What changes when we perform SGD with momentum? What happens when we use minibatch SGD with momentum?

momentum이란 관성 정도로 생각할 수 있는데, weight를 업데이트할 때 내려왔던 방향도 반영을 하자는 것이다. 그래서 local minimum이나 saddle point에서도 그 영역을 탈출할 수 있다.
