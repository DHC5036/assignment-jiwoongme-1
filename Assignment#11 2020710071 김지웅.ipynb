### 1. Propose a strategy to utilize recommender system in healthcare application.

Healthcare분야에서 몇가지 Health recommnedation system을 생각해볼 수 있다. 
첫번째로, Food recommnedation이다.많은 사람들이 부적절한 식습관을 가지고 있기 때문에 적절한 식습관을 위한 음식 혹은 메뉴 추천을 할 수 있다. 또한 어떤 메뉴에 대해 비슷한 영양성분을 가진 다른 메뉴를 추천해 줄 수도 있다. 
두번째로, Drug recommendation이다. 예를 들어, 어떤 약을 처방할 때, 환자의 개인화된 특성에 따라 약의 용량을 달리 할 수 있는데 이를 recommendation system의 도움을 받아 사람마다 더 적절하게 처방할 수 있다.
세번째로, physical activity recommnedation이다. 사용자의 목표나 선호도에 따라 언제 무엇을 먹고 언제 운동을 해야하는지 추천해줌으로써, 환자의 건강 유지를 도울 수 있다.
AI를 활용한 많은 추천 알고리즘들이 많들어지고 있고, 앞으로 더 만들어질테지만 이를 적절하게 평가하고, 알맞게 사용할 수 있어야한다고 생각한다.
