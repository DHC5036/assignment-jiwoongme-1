### 1. What happens to the gradient if you backpropagate through a long sequence?

과거로 가는 과정에서 Gradient 계산이 곱셈 연산으로 이루어져있기 때문에 제대로 계산이 안된다. 그래서 결국 1보다 큰 값들을 계속 곱해주면 발산하고, 1보다 작은 값들을 계속 곱하면 0으로 수렴해서 사라질 것이다.
