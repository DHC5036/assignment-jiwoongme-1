### 1. Why does BERT succeed?

첫번째로 BERT는 양방향성을 지닌다. Bidrectional Encoder Representations from Transformers에서 알수있듯이 Bert는 양방향성을 포함하여 문맥을 더욱 자연스럽게 파악할 수 있다. 이를 통해 똑같은 단어더라도, 앞 뒤 단어의 관계성을 파악할 수 있다.
두번째로 pre-training이 가능하다. 이전에 존재하던 NLP모델은 pre-training이 어려웠기 때문에 특정 task가 존재할 경우 처음부터 학습시켜야하는 단점이 있었다. BERT는 엄청난 수의 wikipedia와 BooksCorpus의 단어를 학습해서 다양한 task에 적용하고, fine-tuning해서 사용할 수 있는것이 큰 장점이다.

### 2. Can we leverage BERT in clinical natural language processing

BERT는 pre-training이 가능하기 때문에 임상 NLP영역에서도 충분히 활용될 수 있다고 생각한다.MIMIC3를 학습시킨 Clinical BERT가 구성되어있고, 관련 연구들이 있다. 일반적인 BERT와 BioBERT, Clinical BERRT, Discharge summary BERT, Bio+Clnical BERT, Bio+Discharge summary BERT 등의 clinical영역의 데이터를 미리 학습시켜 기존의 NLP모델보다 뛰어난 성능을 보일 수 있다.
