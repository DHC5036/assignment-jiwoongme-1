#### Is it efcient to use matrix multiplication to implement convolution operations? Why

Convolution은 이미지의 형상을 무시하지 않고 이미지를 그대로 학습할 수 있게 해주는 수학 행렬 연산이다. 
Flattening함으로써 데이터의 공간적 구조를 무시하지 않고 입력데이터 행렬에 필터 행렬을 합성곱해서 피쳐맵을 구한다.
이때 왼쪽 위에서부터 시작하여 필터의 크기만큼 입력데이터의 크기를 범위를 선택하고, 같은 위치에 있는 원소끼리 성분곱을 하여 그들의 값을 합한다.
